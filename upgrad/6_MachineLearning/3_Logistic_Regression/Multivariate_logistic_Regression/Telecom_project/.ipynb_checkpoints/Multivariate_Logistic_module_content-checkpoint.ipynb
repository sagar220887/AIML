```python
import pandas as pd
import numpy as np
```

## Loading data sets

```python
customer_data = pd.read_csv('customer_data.csv')
customer_data.head()
```

```python
churn_data = pd.read_csv('churn_data.csv')
churn_data.head()
```

```python
internet_data = pd.read_csv('internet_data.csv')
internet_data.head()
```

```python
## merging customer_data with churn_data
df1 = pd.merge(customer_data, churn_data, how="inner", on="customerID")
df1.head()
```

```python
## merging internet_data
telecom = pd.merge(df1, internet_data, how="inner", on="customerID")
telecom.head()
```

```python
telecom.shape
```

```python
telecom.describe()
```

```python
telecom.info()
```

```python
telecom.head(2)
```

```python
for col in telecom.columns:
    if (len(telecom[col].value_counts()) <= 2):
        print(telecom[col].value_counts())
```

```python
binary_columns = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
def binary_map(x):
    return x.map({'Yes': 1, 'No': 0})
    
telecom[binary_columns] = telecom[binary_columns].apply(binary_map)
telecom[binary_columns].head()
```

```python
import numpy as np
import pandas as pd
```

```python
teaCoffee = pd.read_csv('tea_coffee.csv')
teaCoffee
```

```python
X = teaCoffee['Document']
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')

## fit the vect
vect.fit(X)

## get the vocabulary
vect.vocabulary_
```

```python
vect.get_feature_names_out()
```

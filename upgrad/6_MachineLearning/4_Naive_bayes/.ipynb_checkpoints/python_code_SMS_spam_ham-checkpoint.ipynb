```python
import numpy as np
import pandas as pd
```

```python
sms_data = pd.read_csv('smsspamcollection.csv', header=None, names=['Class', 'sms'])
sms_data.head()
```

```python
sms_data.shape
```

```python
sms_data.info()
```

```python
sms_data.isnull().sum()
```

```python
ham_spam = sms_data['Class'].value_counts()
ham_spam
```

```python
spam_ratio = (float(ham_spam[1])/float(ham_spam[0] + ham_spam[1]))*100
print(f'Spam ratio is {round(spam_ratio,2)}')
```

```python
## mapping 0 and 1 to the column Class
sms_data['Class'] = sms_data.Class.map({"spam": 1, "ham": 0}).fillna(-1)
sms_data['Class'].value_counts()
```

```python
sms_data.head()
```

## Assigning X and y with sms and Class values

```python
X = sms_data['sms']
y = sms_data['Class']
```

## splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=100)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

## bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
len(vect.vocabulary_.keys())
```

## Transform both X_train and X_test

```python
X_train_transformed = vect.transform(X_train)
print(X_train_transformed)
```

```python
X_test_transformed = vect.transform(X_test)
print(X_test_transformed)
```

## Building and evaluating Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
## Create object
mnb = MultinomialNB()
```

```python
## fit
mnb.fit(X_train_transformed, y_train)
```

```python
## predict class
y_pred_class = mnb.predict(X_test_transformed)
```

```python
## Predict probabilities
y_pred_proba = mnb.predict_proba(X_test_transformed)
```

## Model Evaluation

```python
from sklearn import metrics
```

```python
## Accuracy
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_class)
TN = confusion[0,0]
TP = confusion[1,1]
FP = confusion[0,1]
FN = confusion[1,0]
```

```python
sensitivity = TP/ float(TP + FN)
print(sensitivity)
```

```python
specificity = TN/ float(TN + FP)
print(specificity)
```

```python
precision = TP/ float(TP + FP)
print(precision)
print(metrics.precision_score(y_test, y_pred_class))
```

```python
recall_score = metrics.recall_score(y_test, y_pred_class)
print(recall_score)
```

```python
f1score = metrics.f1_score(y_test, y_pred_class)
print(f1score)
```

```python
## Creating ROC curve

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
fpr, tpr, threshold = roc_curve(y_test, y_pred_proba[:, 1])
roc_auc = auc(fpr, tpr)
print(roc_auc)
```

```python
## data frame
pd.DataFrame({
    'Threshold': threshold,
    'TPR': tpr,
    'FPR':fpr
}) 
```

```python
%matplotlib inline
plt.ylabel('TPR')
plt.xlabel('FPR')
plt.title('ROC')
plt.plot(fpr,tpr)
```

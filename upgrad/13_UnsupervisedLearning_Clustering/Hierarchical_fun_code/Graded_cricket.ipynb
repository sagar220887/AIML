```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
df = pd.read_csv('Cricket.csv', encoding="ISO-8859-1")
df.head()
```

```python
df_new = df[['Player', 'Ave', 'SR']]
df_new.head()
```

```python
df_new.shape
```

```python
df_new.info()
```

```python
df_param = df_new.drop('Player', axis=1)
df_param.head()
```

```python
## scaling

scaler = StandardScaler();
df_scaled = scaler.fit_transform(df_param)

df_scaled
```

```python
kmean = KMeans(n_clusters=4, max_iter=50)
kmean.fit(df_scaled)
```

```python
kmean_labels = kmean.labels_
kmean_labels
```

```python
df_new['cluster_id'] = kmean_labels
df_new.head()
```

```python
df_new[df_new['cluster_id'] == 1]
```

```python
## Single Linkage

mergings = linkage(df_scaled, method='single', metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
cluster_labels = cut_tree(mergings, n_clusters=4).reshape(-1,)
cluster_labels
```

```python
df_new.head()
```

```python
df_new.drop('cluster_id', axis=1, inplace=True)
```

```python
df_new.head()
```

```python
df_new['cluster_id'] = cluster_labels
df_new.head()
```

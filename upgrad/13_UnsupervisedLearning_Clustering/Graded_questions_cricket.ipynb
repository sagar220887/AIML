```python
string = 'HelloUpgrad'
prevChar = ""
curr_longest = ""
longest = ""

for char in string:
    print('-------------------')
    print('prevChar - ', prevChar)
    print('curr_longest - ', curr_longest)
    print('longest - ', longest)
    
    print('char - ', char)
    
    if (prevChar.lower()<=char.lower()):
        print('------- prevChar<=char')
        curr_longest += char
        print('curr_longest - ', curr_longest)
        if len(curr_longest) > len(longest):
            longest= curr_longest
    else:
        print('-------- prevChar greater than char')
        curr_longest = char
    prevChar = char
print( longest )
```

```python
n= 4

for i in range(0,n):
    print(i*'#', end='')
    total_star = (2*(n-i)) -1
    print(total_star * '*', end='')
    print(i*'#')
```

```python
n= 5
total_chars = (2*n)-1 # 7

for i in range(1,n+1):
    total_stars = (2*i) -1
    total_hash = total_chars - total_stars
    side_hash = int(total_hash/2)
    
    print(side_hash*'#', end='')
    print(total_stars * '*', end='')
    print(side_hash*'#')
```

```python
input_list = [2,1,3,4,1,5,6,1,7]
minimas=[]

for i in range(1,len(input_list)-1):
    prev_num = input_list[i-1]
    next_num = input_list[i+1]
    if((prev_num > input_list[i]) & (next_num > input_list[i])):
        minimas.append(i)
        
print(minimas)
```

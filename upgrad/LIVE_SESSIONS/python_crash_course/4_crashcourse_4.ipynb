## working with csv

```python
df = pd.read_csv('./Python_basics_shared_instructor/course_4/data/titanic_train.csv')
df.head()
```

## BAR PLOT

```python
embarked = df['Embarked'].value_counts()
```

```python
embarked.plot.bar()
```

```python
plt.bar(x=embarked.index, height=embarked.values)
```

```python
plt.bar(x=embarked.index, height=embarked.values, color=['blue','red', 'green'])
```

## Horizontal bar chart

```python
plt.barh(y=embarked.index, width=embarked.values, color=['blue','red', 'green'])
```

```python
embarked_male = df[df['Sex']=='male']['Embarked'].value_counts()
embarked_female = df[df['Sex']=='female']['Embarked'].value_counts()
```

```python
plt.bar(x=embarked_male.index, height=embarked_male.values, color='red')
plt.bar(x=embarked_female.index, height=embarked_female.values, color='blue')
# plt.legends(legends=['male','female'])
```

## HISTOGRAM

```python
ages = df['Age']

plt.hist(ages, bins=20)
plt.xlabel('Age')
plt.ylabel('Passengers')
plt.title('Distribution of passengers age group')
plt.grid(True)
```

## Pie charts

```python
x = df['Sex'].value_counts()
plt.pie(x, labels=['male','female'])
```

```python
plt.pie(x, labels=['male','female'], startangle=90)
```

```python
plt.pie(x, labels=['male','female'], startangle=90, explode=[0.1,0])
```

```python
plt.pie(x, labels=['male','female'], startangle=90, explode=[0.1,0], autopct='%.2f%%')
```

## scatter plot

```python
age = df['Age']
fare = df['Fare']

plt.scatter(x=age, y=fare)
```

```python
age_m = df[df['Sex'] == 'male']['Age']
age_f = df[df['Sex'] == 'female']['Age']

fare_m = df[df['Sex'] == 'male']['Fare']
fare_f = df[df['Sex'] == 'female']['Fare']

plt.scatter(x=age_m, y=fare_m, label='Male')
plt.scatter(x=age_f, y=fare_f, label='Female')
plt.legend()
# plt.legend(['Male', 'Female'])
```

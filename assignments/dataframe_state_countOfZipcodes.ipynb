# 2. Create a dataframe with State names and count of zipcodes as the columns and sort the dataframe in ascending order

```python
df = pd.read_csv('https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv')
df.head()
```

```python
# Groupby state and setting the aggregate function as count
state_groupby = df.groupby(by=(['state'])).count()

# sorting it by zipcode count in ascending order
state_groupby = state_groupby.sort_values('zipcode',ascending=True)

# removing the index 
state_groupby.reset_index(inplace=True)

#Renaming the zipcode column to 'count_of_zipcodes'
state_groupby.rename(columns={'zipcode': 'count_of_zipcodes'}, inplace=True)

# assigning the dataframe with columns only state and count_of_zipcodes
new_df = state_groupby[['state','count_of_zipcodes']]
new_df
```

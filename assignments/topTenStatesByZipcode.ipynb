```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv')
df.head()
```

```python
# Group by the data by states and taking the count
state_groupby = df.groupby(by=(['state'])).count()

# Sorting the data in descending for the column zipcode
df_filt = state_groupby.sort_values('zipcode',ascending=False).head(10)

#Fetching the index from the groupby dataset and storing it in a numpy array
top_ten_states_byzipcode = np.array(df_filt.index)
print(top_ten_states_byzipcode)
```
